```python
df['license'].fillna('NAN',inplace = True)
df['house_rules'].fillna('NAN',inplace = True)
df['host_identity_verified'].fillna('unconfirmed',inplace = True)
df['country'].fillna('United States',inplace = True)
df['country code'].fillna('US',inplace = True)
df['number of reviews'].fillna(0,inplace = True)
df['last review'].fillna('NAN',inplace = True)
df['reviews per month'].fillna('NAN',inplace = True)
df['review rate number'].fillna('NAN',inplace = True)
df['calculated host listings count'].fillna(0,inplace = True)
df['availability 365'].fillna(0,inplace = True)
```

```python
#1) What is the distribution of host identity verification status?
value_counts = df['host_identity_verified'].value_counts()

# Create a pie chart
plt.figure(figsize=(4,4))
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Host Identity Verification')
plt.show()
```

```python
#2) What is the average number of listings per host?
df['calculated host listings count'].mean()
```

```python
#3) Are there hosts who frequently offer instant bookable accommodations?
book = df['instant_bookable'].value_counts()
bookDF = pd.DataFrame(book)
bookDF.columns = ['count']
bookDF
```

```python
#5) What is the most common cancellation policy among hosts?
cancellation_policy = df['cancellation_policy'].value_counts()
cancellation_policyDF = pd.DataFrame(cancellation_policy)
cancellation_policyDF.columns = ['count']
cancellation_policyDF
```

```python
# Create a pie chart
plt.figure(figsize=(4,4))
plt.pie(cancellation_policyDF['count'], labels=cancellation_policyDF.index, autopct='%1.1f%%',colors=['skyblue', 'lightcoral','lightgreen'])
plt.title('Cancellation Policy')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
#Accommodation-related Analysis:
#1)What are the different types of room listings available?
room = df['room type'].value_counts()
roomDF = pd.DataFrame(room)
roomDF
```

```python
room = df['room type'].value_counts()
roomDF = pd.DataFrame(room)
# Create a bar graph with labels
plt.figure(figsize=(8, 6))
ax = roomDF.plot(kind='bar',color = 'lightcoral')
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
```
